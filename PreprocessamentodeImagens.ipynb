# **Normalização e Ajuste de Contraste**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carrega a imagem
image = cv2.imread('imagem2_vc.jpeg', cv2.IMREAD_GRAYSCALE)

# Normalização
normalized_image = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX)

# Ajuste de contraste usando equalização de histograma
equalized_image = cv2.equalizeHist(normalized_image)

# Exibe as imagens usando matplotlib
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(normalized_image, cmap='gray')
plt.title('Imagem Normalizada')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title('Imagem com Contraste Ajustado')
plt.axis('off')

plt.show()


```

# **Filtros de Suavização e Detecção de Bordas:**

```python
import cv2
import matplotlib.pyplot as plt

# Carrega a imagem
image = cv2.imread('imagem3_vc.jpeg', cv2.IMREAD_GRAYSCALE)

# Filtro Gaussiano para suavização
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

# Detecção de bordas com Canny
edges = cv2.Canny(blurred_image, 100, 200)

# Exibe as imagens usando matplotlib
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(blurred_image, cmap='gray')
plt.title('Imagem Suavizada')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title('Bordas Detectadas')
plt.axis('off')

plt.show()

```

# **Transformações Geométricas e Correção de Distorções**

```python

import cv2
import matplotlib.pyplot as plt

# Carrega a imagem
image = cv2.imread('imagem4_vc.jpeg')

# Redimensionamento
resized_image = cv2.resize(image, (300, 300))

# Rotação
(h, w) = image.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_image = cv2.warpAffine(image, M, (w, h))

# Exibe as imagens usando matplotlib
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title('Imagem Redimensionada')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Imagem Rotacionada')
plt.axis('off')

plt.show()
```
